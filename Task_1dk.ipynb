```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)

# Load Dataset

df = pd.read_csv("atlantis_citizens_final.csv")

# Q1. How does average wealth index vary across districts?
q1 = df.groupby("District_Name")["Wealth_Index"].mean().sort_values(ascending=False)
q1

# Q2. What fraction of citizens work outside their home district?
df["Works_Outside"] = df["District_Name"] != df["Work_District"]
q2 = df["Works_Outside"].mean()
q2

# Q3. How does working outside vary by district?
q3 = df.groupby("District_Name")["Works_Outside"].mean().sort_values(ascending=False)
q3

# Q4. Are certain occupations more likely to work outside?
q4 = df.groupby("Occupation")["Works_Outside"].mean().sort_values(ascending=False)
q4

# Q5. How does wealth index differ across occupations?
q5 = df.groupby("Occupation")["Wealth_Index"].mean().sort_values(ascending=False)
q5

sns.boxplot(data=df, x="Occupation", y="Wealth_Index")
plt.xticks(rotation=45)
plt.title("Wealth Index by Occupation")
plt.show()


# Q6. Relationship between house size and life expectancy
q6 = df["House_Size_sq_ft"].corr(df["Life_Expectancy"])
q6

sns.scatterplot(data=df, x="House_Size_sq_ft", y="Life_Expectancy")
plt.title("House Size vs Life Expectancy")
plt.show()

# Q7. How does diet type relate to wealth index?
q7 = df.groupby("Diet_Type")["Wealth_Index"].mean().sort_values(ascending=False)
q7

sns.boxplot(data=df, x="Diet_Type", y="Wealth_Index")
plt.title("Wealth Index vs Diet Type")
plt.show()


# Q8. Does vehicle ownership correspond to higher wealth?
q8 = df.groupby("Vehicle_Owned")["Wealth_Index"].mean()
q8

sns.boxplot(data=df, x="Vehicle_Owned", y="Wealth_Index")
plt.title("Wealth Index vs Vehicle Ownership")
plt.show()

# Q9. How does occupational composition vary by district?
q9 = pd.crosstab(df["District_Name"], df["Occupation"], normalize="index")
q9


# Q10. Does Bio_Hash show clustering with occupation?
q10 = pd.crosstab(df["Bio_Hash"], df["Occupation"], normalize="index")
q10
#Correlation analysis
sns.heatmap(df.corr(numeric_only=True), annot=True)
# Q11. How many citizens are there in each district?
q1 = df["District_Name"].value_counts()
q1
```
